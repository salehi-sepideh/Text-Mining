```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('textClassification.csv')
```

```python
data.head()
```

```python
data.columns = ['Y', 'data']
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data[data['data'].isnull()]
```

```python
data = data.dropna(subset=['data'])
```

```python
data.isnull().sum()
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
len(stopwords)
```

```python
stopwords
```

```python
from nltk.stem import PorterStemmer
stemer=PorterStemmer()
```

```python
dataset=pd.DataFrame(columns=('title_body', 'category'))
for index,row in data.iterrows():
  words=row['data'].split()
  words_filtered=[w for w in words if not w in stopwords]
  words_filtered_stemmed=(stemer.stem(w) for w in words_filtered)
  dataset.loc[index]={'title_body':' '.join(words_filtered_stemmed),
                      'category':row['Y']}
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
vectorizer.fit(dataset['title_body'])
```

```python
x=vectorizer.transform(dataset['title_body'])
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(dataset['category'])
```

```python
le.classes_
```

```python
len(y)
```

```python
np.shape(y)
```

```python
np.shape(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
y_pred=svc.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
