```python
import pandas as pd
import numpy as np
import nltk
import hazm
```

```python
data=pd.read_csv('per.csv')
```

```python
data.head()
```

```python
with open ('stopwords.txt' , encoding='utf8') as stopwords_file:
  stopwords=stopwords_file.readlines()
stopwords=[line.replace('\n','') for line in stopwords]
```

```python
stopwords
```

```python
nltk.download('stopwords')
```

```python
nltk_stopwords=nltk.corpus.stopwords.words('english')
```

```python
stopwords.extend(nltk_stopwords)
```

```python
len(stopwords)
```

```python
stemer=hazm.Stemmer()
```

```python
from hazm import word_tokenize
```

```python
dataset=pd.DataFrame(columns=('title_body','category'))
```

```python
for index,row in data.iterrows():
  title_body=row['Title']+""+row['Body']
  title_body_tokenized=word_tokenize(title_body)
  title_body_tokenized_filtered=[w for w in title_body_tokenized if not w in stopwords]
  title_body_tokenized_filtered_stemmed=[stemer.stem(w) for w in title_body_tokenized_filtered]
  dataset.loc[index]={'title_body':" ".join(title_body_tokenized_filtered_stemmed),
                      'category':row['Category2'].replace('\n','')}
```

```python
dataset.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer()
```

```python
vectorizer.fit(dataset['title_body'])
```

```python
x=vectorizer.transform(dataset['title_body'])
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
y=le.fit_transform(dataset['category'])
```

```python
le.classes_
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.svm import SVC
svmc=SVC()
```

```python
svmc.fit(x_train,y_train)
```

```python
svmc.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred=svmc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
